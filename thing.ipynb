```python
import pandas as pd
```

```python
# Loading the xy file

df = pd.read_csv("sf75.xy", delimiter="\t\t", header = None)
df.head()
```

```python
#Plot the x y coordinates

import matplotlib.pyplot as plt

plt.plot(df[0], df[1])
plt.show()
```

```python
# Renaming the columns to X and Y
df.columns = ["X", "Y"]
```

```python
df.head()
```

```python
plt.plot(df['X'], df['Y'])
plt.show()
```

```python
import math
angle_deg = 5

# Calculate Z which is just z = tan(wedge_angle / 2) * y
df['z'] = df['Y'] * math.tan(math.radians(angle_deg/2))
```

```python
df.head()
```

```python
# save the new file which also has z coordinates now
# this file will be used in the blockMeshDictgen.py file
df.to_csv("sf75withz.csv", index=False)
```
